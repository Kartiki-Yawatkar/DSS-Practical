# Line Chart

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]   # now length = 5
plt.plot(x, y)
plt.title("Line Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Bar Chart

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
a=(1,5,8,6,3,7,9,4)
b=(10,55,20,33,42,85,84,72)

plt.scatter(a,b)
plt.title("scatter plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show
```
